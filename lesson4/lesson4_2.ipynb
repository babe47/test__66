```python
age = int(input("請輸入年齡："))
if age >= 18:
    print("滿18了" + "\n可以考駕照")
else:
    print("未滿18" + f"\n再等{18-age}年，就可以考駕照")
print("The End")
```

```python
age = int(input("請輸入年齡："))
if age < 18:
    print("未滿18" + "\n不可以考駕照")
elif age <= 25:
    print("可以考駕照")
else:
    print("可以考大型重機駕照")
print("The End")
```
