```python
import paho.mqtt.client as mqtt
import json
```

```python
# MQTT Broker è¨­å®š
BROKER = "localhost"
PORT = 1883
TOPIC = "å®¢å»³/æº«åº¦"

# å»ºç«‹ MQTT å®¢æˆ¶ç«¯
client = mqtt.Client(callback_api_version=mqtt.CallbackAPIVersion.VERSION2)

# é€£ç·šå›èª¿
def on_connect(client, userdata, flags, reason_code, properties):
    if reason_code.is_failure:
        print(f"âŒ é€£ç·šå¤±æ•—ï¼ŒéŒ¯èª¤ä»£ç¢¼: {reason_code}")
    else:
        print(f"âœ… æˆåŠŸé€£æ¥åˆ° MQTT Broker: {BROKER}")
        # é€£ç·šæˆåŠŸå¾Œè¨‚é–±ä¸»é¡Œ
        client.subscribe(TOPIC)
        print(f"ğŸ“¡ å·²è¨‚é–±ä¸»é¡Œ: {TOPIC}")

# æ”¶åˆ°è¨Šæ¯å›èª¿
def on_message(client, userdata, msg):
    try:
        payload = msg.payload.decode('utf-8')
        print(f"ğŸ“© æ”¶åˆ°è¨Šæ¯ - ä¸»é¡Œ: {msg.topic}")
        print(f"   å…§å®¹: {payload}")
        
        # å˜—è©¦è§£æ JSON
        try:
            data = json.loads(payload)
            print(f"   è§£æå¾Œçš„è³‡æ–™: {data}")
        except json.JSONDecodeError:
            pass
            
    except Exception as e:
        print(f"âŒ è™•ç†è¨Šæ¯æ™‚ç™¼ç”ŸéŒ¯èª¤: {e}")

# è¨­å®šå›èª¿å‡½æ•¸
client.on_connect = on_connect
client.on_message = on_message

# é€£æ¥ä¸¦é–‹å§‹ç›£è½
print(f"æ­£åœ¨é€£æ¥åˆ° {BROKER}...")
client.connect(BROKER, PORT, 60)

# é–‹å§‹æ°¸ä¹…è¿´åœˆç­‰å¾…è¨Šæ¯ (æŒ‰ Stop æŒ‰éˆ•å¯åœæ­¢)
try:
    print("â³ ç­‰å¾…è¨Šæ¯ä¸­... (æŒ‰ Stop åœæ­¢)")
    client.loop_forever()
except KeyboardInterrupt:
    print("\nğŸ›‘ ç¨‹å¼å·²åœæ­¢")
    client.disconnect()
```
